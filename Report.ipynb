*A Geospatial Assessment of Disparities in Access to Education*

## **Executive Summary**  
This study employs geospatial analysis to investigate disparities in educational access across Nigeria. The findings reveal:

- **Significant regional disparities** in school distribution, particularly between Northern and Southern Nigeria.
- **Students in Northern Nigeria travel 2.5√ó farther** to universities than their Southern counterparts due to a combination of fewer universities, poor road infrastructure, and the geographical spread of institutions.

## **2. Key Findings**  
### **2.1 Distribution of Educational Institutions**  
The study identified three major educational access zones in Nigeria:

| Region | Avg. Distance to University | Key Characteristics |
|---------|----------------------|---------------------|
| North | 385km | Sparse school distribution, low accessibility |
| Middle Belt | 220km | Moderate infrastructure, scattered institutions |
| South | 140km | High concentration of institutions, shorter travel distances |

### **2.2 Travel Distance and Accessibility**  
#### **Key Findings:**  
- **Students in Northern Nigeria travel up to 12 hours** to the nearest university, compared to 3-5 hours in the South, primarily due to fewer universities, poor road networks, and geographical spread.
- **70% of schools in the North are >30km away** from major roads, making daily attendance difficult.
- **Gender disparity**: Female student enrollment is 30% lower in regions with poor road networks.

#### **Road Network Analysis**  
| Route | Dry Season Travel Time | Rainy Season Travel Time | Impact |
|-------|------------|--------------|--------|
| Kano-Sokoto | 5.8 hrs | 9.4 hrs | +62% delay |
| Lagos-Ibadan | 2.0 hrs | 2.5 hrs | +15% delay |

### **2.3 Identification of Underserved Areas**  
Underserved LGAs were identified based on:
1. Low school density (bottom 10% nationally).
2. Distance >300km to the nearest university.
3. Poor road infrastructure (over 75% unpaved roads).

**Priority LGAs for Intervention:**  
- **Sokoto North** (High distance, poor road network)
- **Borno South** (Sparse school distribution)
- **Taraba West** (Limited access to secondary and tertiary education)

## **4. Methodology**  
### **4.1 Data Sources**  
| Data | Records | Use |
|------|---------|-----|
| Primary & Secondary Schools | 30,000+ | Spatial distribution mapping |
| Universities | 270+ | Accessibility analysis |
| Nigerian Road Network | 28,000km | Travel time modeling |

### **4.2 Analytical Techniques**  
1. **Data Cleaning & Validation**:  
   - Removed 18 duplicate university entries.  
   - Corrected GPS coordinate errors in school locations.

2. **Clustering & Spatial Analysis**:  
   - K-Means clustering (k=3) identified educational access zones.  
   - DBSCAN applied for density-based clustering analysis.  
   - Moran‚Äôs I confirmed significant spatial clustering (p<0.001).  
   - Silhouette Score calculated to assess clustering quality (0.683).  
   - Calinski-Harabasz Index used to further validate cluster separation (95.1).  

3. **Network Analysis**:  
   - Estimated travel times using OpenStreetMap road network data.  
   - Measured seasonal delays in school accessibility.

---

## **5. Conclusion**  
This analysis confirms that **geography is a major determinant of educational access in Nigeria.** Key findings include:
1. **The North experiences severe educational access disparities** due to sparse school distribution, poor infrastructure, and vast geographic spread.
2. **Road networks significantly impact school attendance**, especially in rural regions.
3. **Targeted policy interventions** can help bridge regional inequalities and improve nationwide access to education.
